```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

image_data = nib.load('../images/T1.nii.gz').get_fdata()
image_data_v2 = nib.load('../images/T1_V2.nii.gz').get_fdata()
```

```python
mean_value = image_data[image_data > 10].mean()
standard_deviation_value = image_data[image_data > 10].std()


image_data_rescaled = (image_data - mean_value) / (standard_deviation_value)

mean_value = image_data_v2[image_data_v2 > 10].mean()
standard_deviation_value = image_data_v2[image_data_v2 > 10].std()


image_data_rescaled_v2 = (image_data_v2 - mean_value) / (standard_deviation_value)
```

```python
plt.hist(image_data_rescaled_v2[image_data>10].flatten(),100, alpha=0.5)
plt.hist(image_data_rescaled[image_data>10].flatten(),100, alpha=0.5)
```
