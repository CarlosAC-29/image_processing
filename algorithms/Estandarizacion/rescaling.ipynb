```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

image_data = nib.load('../images/T1.nii.gz').get_fdata()
image_data_v2 = nib.load('../images/T1_V2.nii.gz').get_fdata()
```

```python

min_value = image_data.min()
max_value = image_data.max()

image_data_rescaled = (image_data - min_value) / (max_value - min_value)


min_value = image_data_v2.min()
max_value = image_data_v2.max()

image_data_rescaled_2 = (image_data_v2 - min_value) / (max_value - min_value)

```

```python
plt.hist(image_data_rescaled[image_data_rescaled>0.01].flatten(),100, alpha=0.5)
plt.hist(image_data_rescaled_2[image_data_rescaled_2>0.01].flatten(),100, alpha=0.5)
```
